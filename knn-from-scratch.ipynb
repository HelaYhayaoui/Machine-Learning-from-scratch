# k Nearest Neighbors

**Steps**

**1. Calculate the euclidean distance between the new point and every other point.**

**2. select the k distances that are the closest.**

**3. Rank by majority of votes** 

```python
class knn:
        
    def train(self, x,y,k):
        self.k=k
        self.xt=x
        self.yt=y
        
    def predict(self, x):
        labels=[self.pred_one(xi) for xi in x]
        return(labels)
    
    
    def distance (self, qi, pi):
        return np.sqrt(np.sum((qi-pi) ** 2))

    def pred_one(self, xi):
        #distance
        dist = [self.distance(xi, xtrain) for xtrain in self.xt]
        
        #k 
        k_indx= np.argsort(dist)[:self.k] 
        k_labels=[self.yt[i] for i in k_indx]
        
        #vote 
    
        more_frequent= Counter(k_labels).most_common(1)
        return more_frequent[0][0]
        
    
```

```python
def accuracy(yt, yp):
    accuracy = np.sum(yt == yp) / len(yt)
    return accuracy

iris = datasets.load_iris()
features, targets = iris.data, iris.target

xtrain, xtest, ytrain, ytest = train_test_split(features, targets, test_size=0.2, random_state=1234)

k = 5
kn = knn()
kn.train(xtrain, ytrain,k=k)
ypred = kn.predict(xtest)
print("accuracy", accuracy(ytest, ypred))
```
